```python
import pandas as pd

# Ruta al archivo CSV en tu Google Drive
file_path = '/content/drive/MyDrive/Colab Notebooks/Datasets/linkedin-jobs-africa.csv'

# Cargar el conjunto de datos
df = pd.read_csv(file_path)

# Mostrar las primeras filas del conjunto de datos
df.head()
```

```python
# Ejemplos de análisis exploratorio
df['location'].value_counts().plot(kind='bar', figsize=(10, 6), title='Ubicaciones más comunes')
```

# 4. Visualización de los salarios y las empresas:

```python
median_salary = df['salary'].median()
df['salary'].fillna(median_salary, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un gráfico de dispersión
plt.figure(figsize=(12, 8))
sns.scatterplot(x='salary', y='company', data=df, alpha=0.5)
plt.title('Relación entre Salario y Compañía')
plt.xlabel('Salario')
plt.ylabel('Compañía')

# Mostrar el gráfico
plt.show()



```
